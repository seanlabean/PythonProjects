```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

f = open('./2022daily.txt', 'r')
l = f.readlines()
ratings = []
for w in l:
    w = w.strip()
    if w[-3:] == "/10":
        try:
            r = float(w[-6:-3].strip(": "))
        except:
            r = float(w[-5:-3].strip(": "))
        ratings.append(r)
f.close()
print(ratings)

score_counter = {"0.0" : 0, "0.5" : 0, "1.0" : 0, "1.5" : 0,
                 "2.0" : 0, "2.5" : 0, "3.0" : 0, "3.5" : 0,
                 "4.0" : 0, "4.5" : 0, "5.0" : 0, "5.5" : 0,
                 "6.0" : 0, "6.5" : 0, "7.0" : 0, "7.5" : 0,
                 "8.0" : 0, "8.5" : 0, "9.0" : 0, "9.5" : 0,
                 "10.0" : 0
                 }
for s in ratings:
    score_counter[str(s)] += 1
print(score_counter)

plt.figure(figsize=(8,6))
plt.hist(ratings)
plt.show()
```

```python
f = open('./2022daily.txt', 'r')
l = f.readlines()
dStr = []
for w in l:
    w = w.strip()
    if w[-3:] == "/10":
        slash_loc = np.where(np.array([*w]) == '/')[0]
        if len(slash_loc) == 3:
            date_str = w[:slash_loc[1]]
        elif len(slash_loc) == 2:
            date_str = w[:slash_loc[0]+3].strip(':')
        dStr.append(date_str + '/2022')
f.close()
#print(dStr)
```

```python
def date2float(dateStr):
    dateObj = datetime.strptime(dateStr, "%m/%d/%Y")#.date()
    return dateObj.timestamp()
```

```python
dFloat = []
sSpring = "03/21/2022"
sSummer = "06/21/2022"
sFall   = "09/21/2022"
sWinter = "12/21/2022"

winterFloat  = []
winterRating = []
springFloat  = []
springRating = []
summerFloat  = []
summerRating = []
fallFloat    = []
fallRating   = []

for i,d in enumerate(dStr):
    dateFloat = date2float(d)
    dFloat.append(dateFloat)
    
    #if dateFloat < date2float(sSpring) or dateFloat > date2float(sWinter):
    #    winterFloat.append(dateFloat)
    #    winterRating.append(ratings[i])
    if dateFloat >= date2float(sSpring) and dateFloat < date2float(sSummer):
        springFloat.append(dateFloat)
        springRating.append(ratings[i])
    elif dateFloat >= date2float(sSummer) and dateFloat < date2float(sFall):
        summerFloat.append(dateFloat)
        summerRating.append(ratings[i])
    elif dateFloat >= date2float(sFall) and dateFloat <= date2float(sWinter):
        fallFloat.append(dateFloat)
        fallRating.append(ratings[i])
    else:
        winterFloat.append(dateFloat)
        winterRating.append(ratings[i])


```

```python
s=100
plt.figure(figsize=(50,6))
plt.plot(dFloat, ratings)
plt.scatter(winterFloat, winterRating, s=s, color="tab:blue")
plt.scatter(springFloat, springRating, s=s, color="tab:green")
plt.scatter(summerFloat, summerRating, s=s, color="tab:red")
plt.scatter(fallFloat, fallRating, s=s, color="tab:orange")

plt.show()
```
